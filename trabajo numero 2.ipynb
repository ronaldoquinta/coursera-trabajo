```python
import yfinance as yf

# Descargar los datos históricos de las acciones de GME (GameStop)
gme = yf.Ticker("GME")
gme_data = gme.history(period="1y")  # Obtener datos de los últimos 12 meses

# Restablecer el índice
gme_data_reset = gme_data.reset_index()

# Guardar el DataFrame en un archivo CSV
gme_data_reset.to_csv("gme_data.csv", index=False)

# Mostrar las primeras 5 filas
print(gme_data_reset.head())
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Descargar los datos de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Resetear el índice
tesla_data.reset_index(inplace=True)

# Función para graficar
def make_graph(data, title):
    plt.figure(figsize=(14, 6))
    plt.plot(data['Date'], data['Close'], label='Precio de cierre')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Llamar a la función para graficar
make_graph(tesla_data, "Precio histórico de las acciones de Tesla")
```
